```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
df=df.drop(columns=["PassengerId","Cabin","Ticket"], axis=1)
df.head()
```

```python
df["family_size"]=df["SibSp"]+df["Parch"]+1
df=df.drop(columns=["SibSp","Parch"], axis=1)
df.head()
```

```python
df["Embarked"].value_counts()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python

df["Title"]=df["Name"].str.extract("([A-Za-z]+)\.")
df.sample(10)
```

```python
# df["AgeGroup"]=pd.cut(df["Age"], bins=[0, 18, 50, np.inf], labels=["Child", "Adult", "Senior"])
```

```python
df["Embarked"]=df["Embarked"].fillna(df["Embarked"].mode()[0])
df.head()
```

```python
df["Title"].value_counts()
```

```python
count =df["Title"].value_counts()
df["Title"]=df["Title"].replace(count[count<10].index,"Rare")
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=["Survived","Name"]), df["Survived"], test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
tranformer=ColumnTransformer(transformers=[
    ("tnf1",SimpleImputer(),["Age"]),
    ("tnf2",OneHotEncoder(drop="first",sparse_output=False),["Title","Sex"]),
    ("tnf3",OrdinalEncoder(categories=[[1,2,3],["S","C","Q"]]),["Pclass","Embarked"])
],remainder="passthrough")
X_train_new= tranformer.fit_transform(X_train)
X_test_new= tranformer.transform(X_test)
```

```python
print(X_train_new.shape)
print(X_test_new.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model=LogisticRegression()
model.fit(X_train_new,y_train)
y_pred=model.predict(X_test_new)
accuracy_score(y_test,y_pred)
```
